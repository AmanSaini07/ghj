# Importing libraries

```python


import requests                   ##for requesting
from bs4 import BeautifulSoup     ##for scraping from web (web-scraping)
import numpy as np                ## for calculations
import pandas as pd               ##for dataframes 
import seaborn as sns             ##for plotting the graphs 
import matplotlib.pyplot as plt    ##for plotting the graphs 

```

# Input of Website's Url's

```python
url = "https://www.flipkart.com/search?q=apple%20xr&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on#&as=off"
```

# Creating empty list
 for name, price , ratings , specifications so we can store the values in them...

```python
name = []
price = []
ratings = []
specifications = []
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')

data = soup.find_all('div', attrs = {'class':"_4rR01T"})
data
```

# Run a for loop 
Loop for scrape all Data in the web-page and give them class and div tags to scrap a specific thing...


```python
for a in soup.find_all('a',href = True,attrs = {'class':'_1fQZEK'}):
    
    product = a.find('div', attrs = {'class':"_4rR01T"})
    if product is None:
        name.append(np.NaN)
    else:
        name.append(product.text)
    rupees = a.find('div', attrs = {'class':'_30jeq3 _1_WHN1'})
    if rupees is None:
        price.append(np.NaN)
    else:
        price.append(rupees.text)
    stars = a.find('div', attrs = {'class':'_3LWZlK'})
    if stars is None:
        ratings.append(np.NaN)
    else:
        ratings.append(stars.text)
    features = a.find('div', attrs = {'class':'fMghEO'})
    if features is None:
        specifications.append(np.NaN)
    else:
        specifications.append(features.text)
   
    
    
```

# check the length for all variables are equal or not
If they are equal then they make a DataFrame else DataFrame shows the error

```python
print(len(name))
print(len(specifications))
print(len(price))
print(len(ratings))

```

Create a DataFrame so that the user can understand it easily..
